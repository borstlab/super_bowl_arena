```python
import numpy as np
import pylab as plt # matplotlib
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from scipy import signal
import cv2 as cv2
```

```python
Input_im = np.load("center_LED_RF.npy")
Input_im = (128+(Input_im)/np.amax(Input_im)*127).astype("uint8")

pic = np.ones([360,720,3],dtype = "uint8")*127

image = cv2.cvtColor(Input_im, cv2.COLOR_GRAY2RGB)    
resized = cv2.resize(image,dsize=(360, 280), interpolation = cv2.INTER_CUBIC)#cv2.INTER_AREA)

pic[0:280,180:540,:]= resized

fig = plt.figure(figsize=(10,5))
ax = plt.subplot(1, 2, 1)
ax.imshow (pic,extent=[0,360,180,0])
ax.set_xlabel("Azimuth [°]")
ax.set_ylabel("Elevation [°]")
ax.set_title("not corrected")
ax.set_xticks([0,45,90,135,180,225,270,315,360])
ax.set_yticks([0,45,90,135,180])


```

```python

%matplotlib inline
#stim = Stimulus((360,720))
#pic2 = stim.rot_equi_img(pic, pic2, roll=0, pitch=0, yaw=-70)
#pic3 = stim.rot_equi_img(pic2, pic3, roll=0, pitch=25, yaw=0) 

import matplotlib.cm as cm
from matplotlib.colors import Normalize
import seaborn as sns

cmap = sns.diverging_palette(250,10, 85,50,center ="light", as_cmap=True)



pic2 = stim.rot_equi_img(pic, pic, roll=0, pitch=0, yaw=4)
pic3 = stim.rot_equi_img(pic2, pic, roll=0, pitch=-15, yaw=0) 
#pic3 = stim.rot_equi_img(pic3, pic, roll=30, pitch=0, yaw=0) 

def z_score (c):
    mean = np.mean(c)
    std = np.std(c)
    return (c-mean)/std
pic = z_score(pic)
pic3 = z_score(pic3)

fig = plt.figure(figsize=(10,15))

ax = plt.subplot(2, 2, 1)
ax.imshow (pic[:,:,0],extent=[0,360,180,0],cmap=cmap,vmin =-15, vmax = 15)
ax.set_xlabel("Azimuth [°]")
ax.set_ylabel("Elevation [°]")
ax.set_title("spatial RF not corrected")
ax.set_xticks([0,45,90,135,180,225,270,315,360])
ax.set_yticks([0,45,90,135,180])
ax.grid()

ax1 = plt.subplot(2, 2, 2)
ax1.imshow (pic3[:,:,0],extent=[0,360,180,0],cmap=cmap,vmin =-15, vmax = 15)
ax1.set_xlabel("Azimuth [°]")
ax1.set_ylabel("Elevation [°]")
ax1.set_title("spatial RF corrected")
ax1.set_xticks([0,45,90,135,180,225,270,315,360])
ax1.set_yticks([0,45,90,135,180])
ax1.grid()

ax2 = plt.subplot(2, 2, 3)
x = np.linspace(-180,180,720)
avg1 = np.mean(pic3[170:190,:,0],axis =0 )
ax2.plot (x,avg1,label="azi")
y = np.linspace(-90,90,360)
avg2 = np.mean(pic3[:,355:375,0],axis =1 )
ax2.plot (y,avg2,label="ele")
ax2.set_xlabel("size [°]")
ax2.set_ylabel("Intensity[dig]")
ax2.set_title("Crosssections corrected")
ax2.set_xlim(-40,40)
ax2.legend()
ax2.grid()




ax5 = plt.subplot(2, 2, 4)
ax5.imshow (pic3[:,:,0],extent=[-180,180,-90,90],cmap=cmap,vmin =-15, vmax = 15)
ax5.set_xlabel("Azimuth [°]")
ax5.set_ylabel("Elevation [°]")
ax5.set_title("spatial RF corrected")
ax5.grid()

ax5.set_xlim(-30,30)
ax5.set_ylim(-30,30)

plt.tight_layout()
#fig.savefig("Arena_Intensity_new.svg")
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42       
#plt.savefig('Arena_RF_offcenter51_1_1.pdf')
```
